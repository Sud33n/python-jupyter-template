### Exercise 1: Line Plot Customization

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_squared():
    try:
        # Generate x values from 0 to 10
        x = np.linspace(0, 10, 100)
        y = x ** 2
        
        # Create the plot
        plt.figure(figsize=(8, 5))
        plt.plot(x, y, label='y = x^2', color='b', linewidth=2)
        
        # Customization
        plt.title("Line Plot of y = x^2")
        plt.xlabel("X values")
        plt.ylabel("Y values")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.legend()
        
        # Show the plot
        plt.show()
        
    except ImportError as e:
        print(f"Error: {e}. Please ensure numpy and matplotlib are installed.")
    except ValueError as e:
        print(f"Value Error: {e}. Please check the data values being plotted.")
    except plt.error as e:
        print(f"Plotting Error: {e}. There may be an issue with matplotlib.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Call the function to execute the plot
plot_squared()
```
