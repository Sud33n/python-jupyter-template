### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    # Data for plots
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    categories = ['A', 'B', 'C', 'D']
    values = [10, 15, 7, 12]
    scatter_x = np.random.rand(50)
    scatter_y = np.random.rand(50)
    pie_values = [30, 20, 25, 25]
    pie_labels = ['Red', 'Blue', 'Green', 'Yellow']

    # Create figure and subplots
    fig, axs = plt.subplots(2, 2, figsize=(10, 8))

    # Line chart
    axs[0, 0].plot(x, y, label='sin(x)', color='blue')
    axs[0, 0].set_title('Line Chart')
    axs[0, 0].legend()

    # Bar chart
    axs[0, 1].bar(categories, values, color='orange')
    axs[0, 1].set_title('Bar Chart')

    # Scatter plot
    axs[1, 0].scatter(scatter_x, scatter_y, color='green')
    axs[1, 0].set_title('Scatter Plot')

    # Pie chart
    axs[1, 1].pie(pie_values, labels=pie_labels, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'yellow'])
    axs[1, 1].set_title('Pie Chart')

    # Adjust layout
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
